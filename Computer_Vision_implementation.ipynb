```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
#Reading the image 
image=cv2.imread("road(1).jpg")
```

```python
plt.imshow(image)
#cv2.waitkey(0)
```

```python
#Extracting the Height and width of image 
h,w=image.shape[:2]
print('height={} ,width={}'.format(h,w))
```

```python
# Extracting the RGB value
R,G,B=image[100,100]
print("R={},G={},B={}".format(R,G,B))
```

```python
#Selection the Reasion of the image
roi=image[150:500,250:700]
plt.imshow(roi)
#cv2.waitKey(0)
```

```python
# Resizing the image
resize=cv2.resize(image,(700,700))
plt.imshow(resize)
```

```python
#Rotating the image 
#calculate the rotation
ratio=800/w
# create a tuple containg the width and height
dim=(800,int(h*ratio))

#resizing the image
resize_aspect=cv2.resize(image,dim)
plt.imshow(resize_aspect)
#cv2.waitkey(0)
```

```python
#drawing a ractangle
output=image.copy()
ractangle=cv2.rectangle(output,(150,700),(600,400),(255,0,0),2)
plt.imshow(ractangle)
```

```python
# Copying the original image
output = image.copy()

# Adding the text using putText() function
text = cv2.putText(output, 'OpenCV Demo', (200, 550),
                cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 0, 0), 2)
```

```python
plt.imshow(text)
```

```python
cv2.imshow("image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Python code to read image
import cv2

# To read image from disk, we use
# cv2.imread function, in below method,
img = cv2.imread("road(1).jpg", cv2.IMREAD_COLOR)

# Creating GUI window to display an image on screen
# first Parameter is windows title (should be in string format)
# Second Parameter is image array
cv2.imshow("image", img)

# To hold the window on screen, we use cv2.waitKey method
# Once it detected the close input, it will release the control
# To the next line
# First Parameter is for holding screen for specified milliseconds
# It should be positive integer. If 0 pass an parameter, then it will
# hold the screen until user close it.
cv2.waitKey(0)

# It is for removing/deleting created GUI window from screen
# and memory
cv2.destroyAllWindows()
```
